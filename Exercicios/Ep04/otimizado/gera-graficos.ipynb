```python
x = [64, 100, 128, 200, 256, 512, 600, 900, 1024, 2000, 2048, 3000, 4000]
```

## TEMPO DE EXECUÇÃO

```python
tempo_multmatvet = []
tempo_multmatvet_otimizado = []
tempo_multmatmat = []
tempo_multmatmat_otimizado = []

for n in x:
    arquivo = 'ENERGY_'+ str(n) + '.txt'
    with open("resultados/"+arquivo, 'r') as arquivo:
        linhas = arquivo.readlines()
        
        for linha in linhas:
            if "multmatvet:" in linha:
                partes = linha.split(":")
                valor = float(partes[1].strip())
                tempo_multmatvet.append(valor)
            if "multmatvet_otimizado:" in linha:
                partes = linha.split(":")
                valor = float(partes[1].strip())
                tempo_multmatvet_otimizado.append(valor)
            if "multmatmat:" in linha:
                partes = linha.split(":")
                valor = float(partes[1].strip())
                tempo_multmatmat.append(valor)
            if "multmatmat_otimizado:" in linha:
                partes = linha.split(":")
                valor = float(partes[1].strip())
                tempo_multmatmat_otimizado.append(valor)
```

### Gráfico

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 8))

axs[0].plot(x, tempo_multmatvet, label='Não otimizado', linestyle='-', marker='o', color='b')
axs[0].plot(x, tempo_multmatvet_otimizado, label='Otimizado', linestyle='-', marker='o', color='r')
axs[0].legend()

axs[0].set_xlabel('Tamanho (N)')
axs[0].set_ylabel('Tempo (ms)')
axs[0].set_title('Matriz x Vetor (multmatvet)', fontsize=16)

axs[1].plot(x, tempo_multmatmat, label='Não otimizado', linestyle='-', marker='o', color='b')
axs[1].plot(x, tempo_multmatmat_otimizado, label='Otimizado', linestyle='-', marker='o', color='r')
axs[1].legend()

axs[1].set_xlabel('Tamanho (N)')
axs[1].set_ylabel('Tempo (ms)')
axs[1].set_title('Matriz x Matriz (multmatmat)', fontsize=16)

fig.suptitle('Tempo (ms)', fontsize=20)

plt.savefig("graficos/tempo.png")

plt.show()
```
